```python
import numpy as np
from random import randint
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
list1 = []
list2 = []
list3 = []
list4 = []

for i in range(10):
    list1.append(randint(1, 101))
    list2.append(randint(1, 101))
    list3.append(randint(1, 101))
    list4.append(randint(1, 101))

lists = np.array([list1, list2, list3, list4])

plt.plot(lists[0], c="red", marker="s", ls="--", ms=11, label=f"List 1")
plt.plot(lists[1], c="black", marker="o", ls="--", ms=11, label="List 2")
plt.plot(lists[2], c="blue", marker="^", ls="--", ms=11, label="List 3")
plt.plot(lists[3], c="green", marker="1", ls="--", ms=11, label="List 4")

plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.rcParams["figure.figsize"] = 15, 8
plt.locator_params(axis="x", nbins=10)
plt.xticks(list(range(10)), list(range(1,11)))
plt.show()
```
