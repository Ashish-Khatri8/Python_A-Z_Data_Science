```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["figure.figsize"] = 8, 4
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
stats = pd.read_csv("./Demographic-Data.csv")
```

```python
stats.columns = ["CountryName", "CountryCode", "BirthRate", "InternetUsers", "IncomeGroup"]
```

```python
stats.head()
```

```python
# Distribution
vis1 = sns.distplot(stats.InternetUsers)
```

```python
vis1 = sns.distplot(stats.InternetUsers, bins=30)
```

---

```python
# Boxplots
vis2 = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate")
```

---

```python
# Linear Model Plot
vis3 = sns.lmplot(x="InternetUsers", y="BirthRate", data=stats)
```

```python
vis3 = sns.lmplot(x="InternetUsers", y="BirthRate", data=stats, fit_reg=False)
```

```python
vis3 = sns.lmplot(x="InternetUsers", y="BirthRate", data=stats, \
                  fit_reg=False, hue="IncomeGroup", size=8)
```

---

```python
## Keyword arguments in python
```

```python
# Marker size
vis3 = sns.lmplot(x="InternetUsers", y="BirthRate", data=stats, \
                  fit_reg=False, hue="IncomeGroup", size=7,     \
                  scatter_kws={"s": 100})
```
