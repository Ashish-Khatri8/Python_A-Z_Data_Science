```python
## Filtering a dataframe.
```

```python
import pandas as pd
```

```python
stats = pd.read_csv("./Demographic-Data.csv")
```

```python
stats
```

```python
stats.columns = ["CountryName", "CountryCode", "BirthRate", "InternetUsers", "IncomeGroup"]
```

```python
stats
```

---

```python
stats.InternetUsers < 2
```

```python
Filter = stats.InternetUsers < 2
```

```python
stats[Filter]
```

```python
stats[stats.BirthRate > 45]
```

---

```python
Filter1 = stats.InternetUsers < 2
Filter2 = stats.BirthRate > 40
```

```python
# Using Bit-Wise Ampersand &.
stats[Filter1 & Filter2]
```

```python
# stats[stats.InternetUsers < 2 & stats.BirthRate > 40]
# Will give error.
```

```python
stats[(stats.InternetUsers < 2) & (stats.BirthRate > 40)]
```

---

```python
stats[stats.IncomeGroup == "High income"]
```
